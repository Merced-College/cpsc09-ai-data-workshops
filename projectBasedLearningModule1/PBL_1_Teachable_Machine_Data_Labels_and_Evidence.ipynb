## 🧠 Training Settings & What They Mean for Your Model
---

In *Data 8 Chapter 1*, we learn that data science combines **data and computation** to explore patterns and make predictions.

As the textbook explains:

> “Data science is about drawing useful conclusions from data using computation.”

When we adjust training settings in Google Teachable Machine, we are making different **modeling choices** that affect how well the model learns and predicts.

---

### 🔁 Epochs — How Many Times the Model Sees the Data

Each **epoch** is one full pass through all of your training images.

- More epochs give the model more opportunities to adjust.
- Too many epochs can reduce performance on new test images.

---

### 📦 Batch Size — How the Model Groups Data During Learning

**Batch size** determines how many images are processed before the model updates itself.

- Smaller batches may introduce more variation.
- Larger batches may produce smoother updates.

---

### 🚀 Learning Rate — How Big of a Step the Model Takes

The **learning rate** controls how much the model adjusts during each update.

- Too small → learning is slow.
- Too large → learning can become unstable.
- A setting that helps one image may hurt another.

---

## ⚖️ Why This Matters

In Chapter 1, Data 8 shows that data science is not just about running code — it is about understanding how computation interacts with data to produce predictions.

Training settings are part of the modeling process.

Changing the assumptions changes the outcome.

# 🔬 Model Behavior Experiments
---

You are going to change **one setting at a time** in Google Teachable Machine and observe how the training graphs respond.

The goal is not to “find the best setting.”  
The goal is to understand how modeling choices affect results.

| Experiment | What You Change | What You Might Notice in the Graphs | What This Teaches (Data 8 Connection) |
|------------|----------------|--------------------------------------|----------------------------------------|
| 1️⃣ Baseline | Keep default settings | Training and validation accuracy rise gradually | This is your **control condition** — like in an experiment |
| 2️⃣ Increase Epochs | Set epochs very high (e.g., 1500+) | Training accuracy keeps increasing; validation may stop improving or drop | More computation does not always improve prediction |
| 3️⃣ Increase Learning Rate | Raise the learning rate | Accuracy may jump around or behave unstably | Large modeling changes can reduce reliability |
| 4️⃣ Decrease Learning Rate | Lower the learning rate | Slower improvement in accuracy | Tradeoff between speed and stability |
| 5️⃣ Imbalanced Classes | Add many more images to one class | One class performs much better than the other | Unequal sampling affects conclusions |
| 6️⃣ Confounded Data | Use different backgrounds/lighting for each class | High training accuracy but weaker performance on new examples | Confounding variables distort interpretation |

# 📚 Building Our Dataset: Little Women vs Huckleberry Finn (Illustrations)
---

In Data 8 Chapter 1, we analyzed patterns in *Little Women* and *Huckleberry Finn* by examining structure in the text.

Now we will examine structure in the **illustrations**.

Instead of counting words in tables, we will:

   - `test` images (used later in Colab to evaluate the exported model)

This code builds that dataset for you automatically.

---

## 🔍 What the Code Does (In Plain Language)

- Downloads illustrated editions from Project Gutenberg.
- Extracts illustration images.
- Filters out very small decorative images.
- Randomly splits the images into training and testing sets.
- Organizes them into labeled folders.
- Creates a zip file so you can download and use them in GTM.

## ▶️ What You Should Do

```python
# ============================================================
# Data 8 Workshop Dataset Builder (Best GTM success)
# Little Women vs Huckleberry Finn — Illustration Images
# Creates: train/ + test/ folders + ZIP for GTM + Colab evaluation
# Sources: Project Gutenberg HTML (zip) with images
#   - Little Women illustrated edition: https://www.gutenberg.org/cache/epub/37106/pg37106-h.zip
#   - Huck Finn: https://www.gutenberg.org/cache/epub/76/pg76-h.zip
# ============================================================

import os, re, random, zipfile, shutil
import requests
from PIL import Image

SEED = 8
random.seed(SEED)

# How many images per class (adjust for your workshop time)
N_TRAIN_PER_CLASS = 80
N_TEST_PER_CLASS  = 20

OUT_ROOT = "lw_hf_illustrations"
TRAIN_DIR = os.path.join(OUT_ROOT, "train")
TEST_DIR  = os.path.join(OUT_ROOT, "test")

LW_ZIP_URL = "https://www.gutenberg.org/cache/epub/37106/pg37106-h.zip"
HF_ZIP_URL = "https://www.gutenberg.org/cache/epub/76/pg76-h.zip"

session = requests.Session()
session.headers.update({"User-Agent": "MercedCollege-Data8-Workshop/1.0 (Educational Use)"})

def download_file(url, out_path):
    r = session.get(url, timeout=120)
    r.raise_for_status()
    with open(out_path, "wb") as f:
        f.write(r.content)

def unzip_to(zip_path, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    shutil.unpack_archive(zip_path, out_dir)

def collect_images(root_dir):
    exts = (".jpg", ".jpeg", ".png", ".gif")
    paths = []
    for r, _, files in os.walk(root_dir):
        for fn in files:
            if fn.lower().endswith(exts):
                paths.append(os.path.join(r, fn))
    return paths

def is_good_illustration(path, min_dim=150, min_area=150*150):
    # Filter out tiny ornaments/icons
    try:
        with Image.open(path) as im:
            w, h = im.size
            if w < min_dim or h < min_dim:
                return False
            if w*h < min_area:
                return False
        return True
    except Exception:
        return False

def safe_copy(src, dst):
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    with open(src, "rb") as fsrc, open(dst, "wb") as fdst:
        fdst.write(fsrc.read())

def split_train_test(paths, n_train, n_test):
    rng = random.Random(SEED)
    paths = paths[:]
    rng.shuffle(paths)
    return paths[:n_train], paths[n_train:n_train+n_test]

# --- Download zips ---
print("Downloading Gutenberg zips...")
download_file(LW_ZIP_URL, "little_women_images.zip")
download_file(HF_ZIP_URL, "huck_finn_images.zip")

# --- Unzip ---
print("Unzipping...")
unzip_to("little_women_images.zip", "lw_zip")
unzip_to("huck_finn_images.zip", "hf_zip")

# --- Collect + filter images ---
print("Collecting images...")
lw_all = collect_images("lw_zip")
hf_all = collect_images("hf_zip")

lw_imgs = [p for p in lw_all if is_good_illustration(p)]
hf_imgs = [p for p in hf_all if is_good_illustration(p)]

print(f"Little Women usable images: {len(lw_imgs)}")
print(f"Huck Finn usable images:    {len(hf_imgs)}")

need = N_TRAIN_PER_CLASS + N_TEST_PER_CLASS
if len(lw_imgs) < need or len(hf_imgs) < need:
    raise ValueError(
        f"Not enough usable images. Need {need} per class. "
        f"Got LW={len(lw_imgs)}, HF={len(hf_imgs)}. "
        f"Try lowering N_TRAIN_PER_CLASS/N_TEST_PER_CLASS."
    )

# --- Prepare folders ---
for base in [TRAIN_DIR, TEST_DIR]:
    for cls in ["little_women", "huck_finn"]:
        os.makedirs(os.path.join(base, cls), exist_ok=True)

lw_train, lw_test = split_train_test(lw_imgs, N_TRAIN_PER_CLASS, N_TEST_PER_CLASS)
hf_train, hf_test = split_train_test(hf_imgs, N_TRAIN_PER_CLASS, N_TEST_PER_CLASS)

# --- Copy into dataset structure ---
print("Building dataset folders...")
for i, p in enumerate(lw_train):
    safe_copy(p, os.path.join(TRAIN_DIR, "little_women", f"lw_{i:04d}{os.path.splitext(p)[1].lower()}"))
for i, p in enumerate(lw_test):
    safe_copy(p, os.path.join(TEST_DIR, "little_women", f"lw_test_{i:04d}{os.path.splitext(p)[1].lower()}"))

for i, p in enumerate(hf_train):
    safe_copy(p, os.path.join(TRAIN_DIR, "huck_finn", f"hf_{i:04d}{os.path.splitext(p)[1].lower()}"))
for i, p in enumerate(hf_test):
    safe_copy(p, os.path.join(TEST_DIR, "huck_finn", f"hf_test_{i:04d}{os.path.splitext(p)[1].lower()}"))

# --- Zip it ---
zip_name = "lw_hf_illustrations_dataset.zip"
print("Zipping...")
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for root, _, files in os.walk(OUT_ROOT):
        for f in files:
            full = os.path.join(root, f)
            z.write(full, arcname=full)

print("\n✅ Done!")
print("Zip file:", zip_name)
print("Train folders:", f"{OUT_ROOT}/train/little_women  and  {OUT_ROOT}/train/huck_finn")
print("Test folders: ", f"{OUT_ROOT}/test/little_women   and  {OUT_ROOT}/test/huck_finn")
```

```python
import tensorflow as tf

# Patch: ignore 'groups' arg in DepthwiseConv2D configs during deserialization
from tensorflow.keras.layers import DepthwiseConv2D

class PatchedDepthwiseConv2D(DepthwiseConv2D):
    @classmethod
    def from_config(cls, config):
        config.pop("groups", None)  # remove unsupported arg
        return super().from_config(config)

custom_objects = {"DepthwiseConv2D": PatchedDepthwiseConv2D}

model = tf.keras.models.load_model(
    "keras_model.h5",
    custom_objects=custom_objects,
    compile=False
)

with open("labels.txt", "r", encoding="utf-8") as f:
    labels = [line.strip() for line in f if line.strip()]

print("Loaded model OK.")
print("Labels:", labels)
```

## 📊 What Happens When You Run These Cells?

When you run the next cells, Python is:

- Comparing the model’s predicted labels to the true labels
- Counting how often the model was correct
- Counting where the model made mistakes
- Organizing those results into a confusion table

A **confusion matrix** helps us see:

- How accurate the model is
- Which classes get confused
- Where bias or imbalance might appear

Run the next cells.

---




```python
import os
import numpy as np
import pandas as pd
from PIL import Image

IMG_SIZE = (224, 224)

def preprocess(path):
    img = Image.open(path).convert("RGB")
    img = img.resize(IMG_SIZE)
    arr = np.array(img).astype(np.float32) / 255.0
    return np.expand_dims(arr, axis=0)

TEST_BASE = "lw_hf_illustrations/train"  # change to /test if evaluating test set

results = []

for actual_class in ["little_women", "huck_finn"]:
    folder = os.path.join("lw_hf_illustrations/test", actual_class)
    for file in os.listdir(folder):
        if file.lower().endswith((".jpg", ".jpeg", ".png")):
            path = os.path.join(folder, file)
            x = preprocess(path)
            probs = model.predict(x, verbose=0)[0]
            pred_idx = int(np.argmax(probs))
            pred_label = labels[pred_idx]
            confidence = float(np.max(probs))

            results.append({
                "filename": file,
                "actual": actual_class,
                "predicted": pred_label,
                "confidence": confidence
            })

df = pd.DataFrame(results)
df.head()
```

The above code should output a table.  What does that table show? Run the next code cell to "clean up" that table.

```python
def clean_label(s):
    return s.split(" ", 1)[1] if " " in s else s

df["predicted_clean"] = df["predicted"].apply(clean_label)
df.head()
```

Now to compute the accuracy...run the code in the next cell.

```python
df["correct"] = df["actual"] == df["predicted_clean"]

accuracy = df["correct"].mean()
accuracy
```

```python
pd.crosstab(df["actual"], df["predicted_clean"])
```

# 🔍 Causality Meets Machine Learning
---

## ⚖️ The Confounding Factor Problem

In Week 1 of Data 8, we learned:

To establish **causality**, treatment and control groups must be similar in every way —  
except for the treatment.

If other variables differ, we introduce a **confounding factor**.

---

## 🤖 What Does That Have To Do With Our Model?

Think about the two classes you created.

Now ask:

- Did one class have brighter lighting?
- Was one object always closer to the camera?
- Did one class mostly appear on a desk and the other in someone’s hand?
- Was the background more cluttered for one class?

If something besides the intended label consistently differs between classes,  
then that difference becomes a **confounder**.

Your model may not be learning the intended feature —  
it may be learning the unintended pattern.

---

## 🔬 Data 8 Parallel

| Data 8 (Causality) | Machine Learning |
|--------------------|------------------|
| Groups must differ only in treatment | Classes should differ only in the intended feature |
| Confounders distort causal conclusions | Background patterns distort predictions |
| Similar groups → cleaner inference | Controlled datasets → cleaner classification |

---

## 💡 Big Idea

In causal inference, we control variables to isolate causes.  
In machine learning, we control variables to isolate signal from noise.

If we don’t control for confounding factors,  
our model may confidently predict the wrong thing.

```python
# ============================================================
# 📚 DATA 8 WORKSHOP – CLASSIC vs MODERN BOOK COVERS (ZIP BUILDER)
# Reliable source: Open Library Covers API (avoids Wikimedia 403 issues)
# Output: gTM_bookcovers_dataset.zip
# ============================================================

import os, re, random, zipfile, requests
from urllib.parse import quote

BASE_DIR = "bookcovers_dataset"
TRAIN_RATIO = 0.8
SEED = 8
N_PER_CLASS_TARGET = 80   # change to 50 if you want faster
IMG_SIZE = "L"            # S, M, or L

SESSION = requests.Session()
SESSION.headers.update({
    "User-Agent": "MercedCollege-Data8-Workshop/1.0 (Educational Use; contact: kanemoto.k@mccd.edu)"
})

# ---- Title lists (edit freely) ----
classic_titles = [
    "Little Women", "Pride and Prejudice", "Jane Eyre", "Wuthering Heights",
    "The Adventures of Tom Sawyer", "Adventures of Huckleberry Finn",
    "Moby-Dick", "The Picture of Dorian Gray", "Frankenstein",
    "Dracula", "The Time Machine", "A Tale of Two Cities",
    "Great Expectations", "Oliver Twist", "The Scarlet Letter",
    "The Adventures of Sherlock Holmes", "The Count of Monte Cristo",
    "Les Misérables", "The Odyssey", "Jules Verne"
]

modern_titles = [
    "The Hunger Games", "Harry Potter and the Sorcerer's Stone", "The Fault in Our Stars",
    "Gone Girl", "The Girl on the Train", "The Da Vinci Code",
    "The Martian", "The Night Circus", "The Help",
    "The Kite Runner", "Life of Pi", "The Road",
    "The Goldfinch", "The Handmaid's Tale", "The Curious Incident of the Dog in the Night-Time",
    "The Book Thief", "A Man Called Ove", "Where the Crawdads Sing",
    "Big Little Lies", "The Alchemist"
]

def safe_name(s):
    s = re.sub(r'[\\/:"*?<>|]+', '', s)
    return s.replace(" ", "_")

def search_openlibrary(title, limit=50):
    url = f"https://openlibrary.org/search.json?title={quote(title)}&limit={limit}"
    r = SESSION.get(url, timeout=60)
    r.raise_for_status()
    return r.json().get("docs", [])

def cover_url(cover_id, size="L"):
    return f"https://covers.openlibrary.org/b/id/{cover_id}-{size}.jpg"

def download_image(url, out_path):
    r = SESSION.get(url, timeout=60)
    r.raise_for_status()
    # Skip “not found” tiny placeholder responses
    if r.headers.get("content-type","").lower().startswith("text/"):
        raise ValueError("Not an image response")
    with open(out_path, "wb") as f:
        f.write(r.content)

def gather_covers_for_titles(titles, target_count, label):
    # Collect unique cover IDs from searches
    cover_ids = []
    seen = set()
    for t in titles:
        for d in search_openlibrary(t, limit=60):
            cid = d.get("cover_i")
            if cid and cid not in seen:
                seen.add(cid)
                cover_ids.append(cid)

    random.Random(SEED).shuffle(cover_ids)

    out_dir = os.path.join("tmp_download", label)
    os.makedirs(out_dir, exist_ok=True)

    downloaded, failed = 0, 0
    for cid in cover_ids:
        if downloaded >= target_count:
            break
        url = cover_url(cid, IMG_SIZE)
        out_path = os.path.join(out_dir, f"{label}_{cid}.jpg")
        try:
            download_image(url, out_path)
            downloaded += 1
        except Exception:
            failed += 1

    print(f"{label}: downloaded {downloaded} covers (failed {failed})")
    return out_dir

def split_train_test(paths, train_ratio=0.8):
    rng = random.Random(SEED)
    paths = paths[:]
    rng.shuffle(paths)
    cut = int(len(paths) * train_ratio)
    return paths[:cut], paths[cut:]

def copy_into(paths, cls, split):
    os.makedirs(os.path.join(BASE_DIR, cls, split), exist_ok=True)
    for p in paths:
        dest = os.path.join(BASE_DIR, cls, split, os.path.basename(p))
        with open(p, "rb") as src, open(dest, "wb") as dst:
            dst.write(src.read())

# ---- Download covers ----
classic_dir = gather_covers_for_titles(classic_titles, N_PER_CLASS_TARGET, "classic")
modern_dir  = gather_covers_for_titles(modern_titles,  N_PER_CLASS_TARGET, "modern")

classic_imgs = [os.path.join(classic_dir, f) for f in os.listdir(classic_dir) if f.lower().endswith(".jpg")]
modern_imgs  = [os.path.join(modern_dir,  f) for f in os.listdir(modern_dir)  if f.lower().endswith(".jpg")]

c_train, c_test = split_train_test(classic_imgs, TRAIN_RATIO)
m_train, m_test = split_train_test(modern_imgs,  TRAIN_RATIO)

# ---- Build dataset folders ----
copy_into(c_train, "classic", "train")
copy_into(c_test,  "classic", "test")
copy_into(m_train, "modern",  "train")
copy_into(m_test,  "modern",  "test")

# ---- Zip it ----
zip_name = "gTM_bookcovers_dataset.zip"
with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
    for root, _, files in os.walk(BASE_DIR):
        for f in files:
            full = os.path.join(root, f)
            z.write(full, arcname=full)

print("\n✅ Dataset ready:", zip_name)
print("✅ Contents:", f"{BASE_DIR}/classic/train, classic/test, modern/train, modern/test")
```
